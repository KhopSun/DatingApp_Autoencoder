```python
import json

# Load user data
user_df = pd.read_csv("GeneratedUsers.csv")

# Define additional attribute pools
faculties = [
    "Faculty of Engineering", "Faculty of Science", "Faculty of Arts",
    "Faculty of Architecture", "Faculty of Commerce and Accountancy",
    "Faculty of Political Science", "Faculty of Communication Arts",
    "Faculty of Medicine", "Faculty of Nursing", "Faculty of Fine and Applied Arts"
]

interests_pool = [
    "Travel", "Movies & Shows", "Photography", "Grabbing a drink", "Music", "Sports", "Gym", "Cafe",
    "Nature", "Reading", "Video Games", "Food", "Pets", "Art", "Cooking", "Dancing"
]

personality_pool = [
    "Cheerful", "Reliable", "Active", "Caring", "Introverted", "Organized",
    "Creative", "Thoughtful", "Confident", "Playful", "Ambitious", "Calm", "Chill"
]

values_pool = [
    "Honesty and Trust", "Self-Growth", "Having fun", "Independence and Balance",
    "Meaningful Conversation", "Emotional Support and Empathy", "Physical Affection"
]

relationship_goals = [
    "Long-term partner", "Long-term but casual OK", "Casual but long-term OK",
    "Casual Fun", "New Friends", "Still figuring it out"
]

# Generate profiles
profiles = []
for i, row in user_df.iterrows():
    profile_id = str(uuid.uuid4())
    user_id = row["user_id"]
    interests = random.sample(interests_pool, random.randint(2, 5))
    looking_for = random.choice(relationship_goals)
    faculty = random.choice(faculties)

    # Fake location data
    latitude = 13.73 + random.random() * 0.02
    longitude = 100.52 + random.random() * 0.02
    location = {
        "coords": {
            "latitude": latitude,
            "longitude": longitude,
            "altitude": random.uniform(0, 20),
            "accuracy": random.uniform(20, 60),
            "altitudeAccuracy": None,
            "heading": None,
            "speed": None
        },
        "timestamp": int((datetime.now() - timedelta(days=random.randint(0, 5))).timestamp() * 1000)
    }

    # Filter preferences
    filter_preferences = {
        "1": random.sample(interests_pool, random.randint(1, 3)),
        "2": random.sample(personality_pool, random.randint(0, 2)),
        "3": random.sample(values_pool, random.randint(1, 3))
    }

    profiles.append({
        "profile_id": profile_id,
        "user_id": user_id,
        "interests": json.dumps(["Female"]),
        "location": json.dumps(location),
        "filter_preferences": json.dumps(filter_preferences),
        "faculty": faculty,
        "looking_for": looking_for
    })

# Create DataFrame and save to CSV
profile_df = pd.DataFrame(profiles)
profile_df.to_csv("GeneratedProfiles.csv", index=False)
```
